Automated classification of heart disease is an important task that can assist doctors and technologist in thier screeing process. Recently, various ML models have been adopted or adopted in order to complete the task.

```python
import pandas as pd
```

```python
data = pd.read_csv("heart.csv")
```

```python
data
```

```python
X = data.drop('output',axis = 1)
```

```python
X
```

```python
y = data["output"]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(X,y,test_size=0.2,random_state=50)
```

```python
xtrain.shape
```

```python
xtest.shape
```

```python
from tensorflow.keras.models import Sequential
```

```python
model = Sequential() #BRAIN WITHOUT NEURON
```

```python
from tensorflow.keras.layers import Dense
```

```python
model.add(Dense(100, activation="relu",input_shape = (13,))) #input layer
```

```python
model.add(Dense(200, activation="relu")) #hidden layer
model.add(Dense(300, activation="relu")) #hidden layer
model.add(Dense(200, activation="relu")) #hidden layer
model.add(Dense(400, activation="relu")) #hidden layer
```

```python
model.add(Dense(1, activation="sigmoid"))
```

```python
model.compile(optimizer="adam", loss = "binary_crossentropy", metrics = ["accuracy"])
```

```python
model.fit(xtrain,ytrain,epochs = 100, batch_size=10)
```

```python
ypred = model.predict(xtest)
```

```python
ypred
```

```python
ypred_transformed = []
for i in ypred:
    if i >= 0.5:
        ypred_transformed.append([1])
    else:
        ypred_transformed.append([0])
```

```python
ypred_transformed 
```

```python
import numpy as np
ypred_transformed = np.array(ypred_transformed)
```

```python
ypred_transformed
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(ypred_transformed,ytest)
```
